```python
# import
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import  datasets, model_selection, linear_model, metrics, preprocessing
from sklearn.linear_model import LinearRegression, SGDRegressor, Ridge, LogisticRegression
%matplotlib inline
```

```python
# 2. Data Loading
df = pd.read_csv("datasets/mushrooms.csv")
# print(df.info())
print(df.describe())
```

```python
# print(df.isna())
print(df.duplicated().sum())
```

```python
# Spilit the feateurs and lalbes
label = df.pop('class')
label.head()
```

```python
label_encoder = preprocessing.LabelEncoder()
label_encoder.fit(label)
label_tranformed = label_encoder.transform(label)
```

```python
# (B) for features
ordinal_encoder = preprocessing.OrdinalEncoder()
ordinal_encoder.fit(df)
featurs_transformed = ordinal_encoder.transform(df)
```

```python
# Put the data back to DataFrame
feature_df = pd.DataFrame(featurs_transformed, columns=df.columns)
```

```python
# 5. Data splitting
x_train, x_test, y_train, y_test = model_selection.train_test_split(feature_df, label_tranformed, train_size=100)
```

```python
# 6. Model development
model = LogisticRegression(solver='liblinear')
model.fit(x_train, y_train)
```

```python
# 7. Model evaluation
training_prdiction = model.predict(x_train)
testing_prediction = model.predict(x_test)

print("Training evaluation:")
print("Training loss: ", metrics.log_loss(y_train, training_prdiction))
print(metrics.classification_report(y_train, training_prdiction))

print("Testing evaluation:")
print("Testing loss: ", metrics.log_loss(y_test, testing_prediction))
print(metrics.classification_report(y_test, testing_prediction))
```
